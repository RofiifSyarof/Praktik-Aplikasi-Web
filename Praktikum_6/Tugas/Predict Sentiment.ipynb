# **Prediksi Sentiment per kalimat**

```python
import re

original_text = '''
Masakan ini ngga enak cuih.
'''

def cleansing(sent):
    # Mengubah kata menjadi huruf kecil
    string = sent.lower()
    # Menghapus emoticon dan tanda baca
    string = re.sub(r'[^a-zA-Z0-9]', ' ', string)
    # Menghapus RT
    string = re.sub('rt', ' ', string)
    return string

# cleansing
text = cleansing(original_text)
```

```python
# feature extraction
text_feature = feature_bow.transform([text])
```

```python
# predict
print('Text: ', str(original_text))
print('Sentiment NB: ', model_nb.predict(text_feature)[0])
print('Sentiment NN: ', model_nn.predict(text_feature)[0])
```

# **Predict Sentiment dari Kumpulan Kalimat**

```python
import pandas as pd

data = pd.read_csv('drive/My Drive/Kuliah Semester 5/Aplikasi Web/Praktikum 6/data_tweet_prabowo.csv')
data.head()
```

```python
print(data)
```

```python
data = data.rename(columns={'full_text': 'Tweet'})
data.head()
```

```python
import re

def cleansing(sent):
    # Mengubah kata menjadi huruf kecil
    string = sent.lower()
    # Menghapus emoticon dan tanda baca
    string = re.sub(r'[^a-zA-Z0-9]', ' ', string)
    # Menghapus RT
    string = re.sub('rt', ' ', string)
    return string
```

```python
def predict_sentiment(sent):
    # cleansing
    text = cleansing(str(sent))
    # feature extraction
    text_feature = feature_bow.transform([text])
    # predict
    return model_nb.predict(text_feature)[0]
```

```python
data['Tweet'] = data.Tweet.apply(cleansing)
data.head()
```

```python
data['predicted_sentiment'] = data.Tweet.apply(predict_sentiment)
data.head()
```

```python
print(data)
```

```python
print(data.predicted_sentiment.value_counts())
```

# Cleansing data hapus spasi berlebih

```python
# cleansing data hapus spasi berlebih
import re

def cleansing(sent):
    # Menghapus spasi berlebih
    string = re.sub(' +', ' ', sent)
    return string
```

```python
data['Tweet'] = data.Tweet.apply(cleansing)
data.head()
```

```python
# cleansing url dalam data
import re

def cleansing(sent):
    # Menghapus url
    string = re.sub(r'http\S+', '', sent)
    return string
```

```python
data['Tweet'] = data.Tweet.apply(cleansing)
data.head()
```
